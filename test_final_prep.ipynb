```python
import pandas as pd
df = pd.read_csv("test.csv")
```

```python
df_features = pd.read_csv("features.csv")
```

```python
df_features.head()
```

```python
df["Temperature"] = [0]*len(df)
df["Fuel_Price"] = [0]*len(df)
df["Fuel_Price"] = [0]*len(df)
df["CPI"] = [0]*len(df)
df["Unemployment"] = [0]*len(df)
```

```python
df.head()
```

```python
for i in range(len(df)):
    M = df_features[df_features["Date"] == df["Date"][i]]
    K = M[M["Store"] == df["Store"][i]]
    df["Temperature"][i] = list(K["Temperature"])[0]
    df["Fuel_Price"][i] = list(K["Fuel_Price"])[0]
    df["CPI"][i] = list(K["CPI"])[0]
    df["Unemployment"][i] = list(K["Unemployment"])[0]
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
week = []
for i in range(len(df)) :
    week.append(df["Date"][i].week)
```

```python
df["week"] = week
```

```python
month = []
day = []
year = []
for i in range(len(df)) :
    month.append(df["Date"][i].month)
    day.append(df["Date"][i].day)
    year.append(df["Date"][i].year)
```

```python
df["month"] = month
df["day"] = day
df["year"] = year
```
