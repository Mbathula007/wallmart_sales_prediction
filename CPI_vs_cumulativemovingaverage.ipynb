```python
import numpy as np
y_cpi = np.load("cycle1/y_cpi.npy")
y_cum = np.load("cycle1/y_cumilative.npy")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.plot([i for i in range(48)],y_cpi/max(y_cpi),color = 'g',label = "CPI")
plt.plot([i for i in range(48)],y_cum/max(y_cum),color = 'r',label = "CUM")
plt.legend()
```

```python
M = []
for i in range(1,len(y_cum)+1) :
    M.append(sum(y_cum[:i])/i)
```
