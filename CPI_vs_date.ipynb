```python
import pandas as pd
```

```python
df = pd.read_csv("features.csv")
```

```python
df.head()
```

```python
df["MarkDown1"].unique()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df = df.sort_values("Date")
```

```python
K = []
for i in range(len(df)):
    K.append(df["Date"][i].year)
```

```python
df["year"] = K
```

```python
df.columns
```

```python
temp_df = df[df["Store"] == 1]
temp_df = temp_df[temp_df["year"] == 2010]
```

```python
T = []
for i in range(len(temp_df)):
    T.append(i)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,10))
plt.plot(T,temp_df["CPI"],color = "g",label = "CPI")
plt.legend()
```

```python
import numpy as np
np.save("cycle1/y_cpi",np.array(list(temp_df["CPI"])))
```
